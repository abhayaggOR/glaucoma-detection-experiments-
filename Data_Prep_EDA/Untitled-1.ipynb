## data prep 

```python
import pandas as pd
import os

# --------------------------------------------------
# PATHS
# --------------------------------------------------
input_csv = "/home/abhay/chaksu/20123135/Train/6.0_Glaucoma_Decision/Glaucoma_Decision_Comparison_Bosch_majority.csv"

output_csv = "/home/abhay/chaksu/20123135/Train/6.0_Glaucoma_Decision/Glaucoma_Decision_Comparison_Bosch_majority_fixed.csv"

# --------------------------------------------------
# LOAD CSV
# --------------------------------------------------
df = pd.read_csv(input_csv)

print("CSV loaded successfully")
print("\nOriginal Images column (sample):")
print(df["Images"].head())

# --------------------------------------------------
# IMAGE NAME FIX (CASE-SENSITIVE)
# --------------------------------------------------
def fix_image_name(image_name):
    """
    Examples:
    Image101.jpg-Image101-1.jpg       -> Image101.JPG
    P10_Image1.jpg-P10_Image1-1.jpg   -> P10_Image1.JPG
    """
    first_part = str(image_name).split("-")[0]      # Image101.jpg
    base_name = os.path.splitext(first_part)[0]     # Image101
    return base_name + ".JPG"                        # ONLY extension uppercase

df["Images"] = df["Images"].apply(fix_image_name)

# --------------------------------------------------
# SAVE NEW CSV (ORIGINAL UNTOUCHED)
# --------------------------------------------------
df.to_csv(output_csv, index=False)

print("\n✅ Images column modified correctly")
print("Saved at:", output_csv)

print("\nUpdated Images column (sample):")
print(df["Images"].head())
```

```python
import pandas as pd
import re

# --------------------------------------------------
# PATHS
# --------------------------------------------------
input_csv = "/home/abhay/chaksu/20123135/Train/6.0_Glaucoma_Decision/bosch_fixed_without_p10_image2.csv"

output_csv = "/home/abhay/chaksu/20123135/Train/6.0_Glaucoma_Decision/bosch_upto_p10_only.csv"

# --------------------------------------------------
# LOAD CSV
# --------------------------------------------------
df = pd.read_csv(input_csv)

print("Rows before filtering:", len(df))

# --------------------------------------------------
# FILTER FUNCTION
# --------------------------------------------------
def keep_row(image_name: str) -> bool:
    """
    Keep:
    - All non-P images (e.g. Image101.JPG)
    - P1 to P10 images
    Remove:
    - P11 onwards
    """
    match = re.match(r"^P(\d+)_", image_name)
    if match:
        patient_num = int(match.group(1))
        return patient_num <= 10
    return True  # keep non-P images

# Apply filter
df_filtered = df[df["Images"].apply(keep_row)]

# --------------------------------------------------
# SAVE
# --------------------------------------------------
df_filtered.to_csv(output_csv, index=False)

print("Rows after filtering:", len(df_filtered))
print("Saved at:", output_csv)
```

## forus data prep 

```python
import pandas as pd
import os

# --------------------------------------------------
# PATHS
# --------------------------------------------------
input_csv = "/home/abhay/chaksu/20123135/Train/6.0_Glaucoma_Decision/Glaucoma_Decision_Comparison_Forus_majority.csv"

output_csv = "/home/abhay/chaksu/20123135/Train/6.0_Glaucoma_Decision/Glaucoma_Decision_Comparison_Forus_majority_fixed.csv"

# --------------------------------------------------
# LOAD CSV
# --------------------------------------------------
df = pd.read_csv(input_csv)

print("CSV loaded successfully")
print("\nOriginal Images column (sample):")
print(df["Images"].head())

# --------------------------------------------------
# IMAGE NAME TRANSFORMATION
# --------------------------------------------------
def fix_image_name(image_name):
    """
    Examples:
    12.jpg-12-1.jpg -> 12.png
    1.jpg-1-1.jpg   -> 1.png
    """
    first_part = str(image_name).split("-")[0]      # e.g. 12.jpg
    base_name = os.path.splitext(first_part)[0]     # e.g. 12
    return base_name + ".png"

df["Images"] = df["Images"].apply(fix_image_name)

# --------------------------------------------------
# SAVE NEW CSV (ORIGINAL UNTOUCHED)
# --------------------------------------------------
df.to_csv(output_csv, index=False)

print("\n✅ Images column modified correctly")
print("Saved at:", output_csv)

print("\nUpdated Images column (sample):")
print(df["Images"].head())
```

```python
import pandas as pd

# --------------------------------------------------
# LOAD CSV
# --------------------------------------------------
csv_path = "/home/abhay/chaksu/20123135/Train/6.0_Glaucoma_Decision/Glaucoma_Decision_Comparison_Forus_majority_fixed.csv"

df = pd.read_csv(csv_path)

# --------------------------------------------------
# EXTRACT NUMERIC PART FROM IMAGE NAME
# --------------------------------------------------
# Example: "12.png" -> 12
df["image_num"] = df["Images"].str.replace(".png", "", regex=False).astype(int)

# --------------------------------------------------
# SORT NUMERICALLY
# --------------------------------------------------
df = df.sort_values(by="image_num").reset_index(drop=True)

# --------------------------------------------------
# DROP HELPER COLUMN
# --------------------------------------------------
df = df.drop(columns=["image_num"])

# --------------------------------------------------
# SAVE SORTED CSV
# --------------------------------------------------
output_csv = "/home/abhay/chaksu/20123135/Train/6.0_Glaucoma_Decision/Glaucoma_Decision_Comparison_Forus_majority_sorted.csv"
df.to_csv(output_csv, index=False)

print("✅ CSV sorted numerically by image index")
print("Saved at:", output_csv)
```
